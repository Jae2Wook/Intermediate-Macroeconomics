## Preliminaries: Putting the ISLM Curve into Python

The IS curve is a simple linear function! However, incorporating the formula $\tilde{Y_t}=\bar{a}-\bar{b}(R_t-\bar{r})$ into Python can be a little tricky because $R_t$ is the dependent variable, not the independent variable when we are graphing it on the Carthesian plane. If I was doing this homework, these are the steps I would take to graph the IS curve:

* Solve for $R_t$ in this equation $\tilde{Y_t}=\bar{a}-\bar{b}(R_t-\bar{r})$.
* Create an array that contains the range of $\tilde{Y}_t$: $\left(-\bar{a}-\bar{b}\bar{r}, \;\bar{a}+\bar{b}\bar{r}\right)$.  
* Use the formula mentioned in bulletpoint 1 to calculate the corresponding values for $R_t$.
* Graph it! Use the appropriate labeling of values of the x-axis so it looks right.

R_t = (a - Y_st) / b + r\
Y_st = [-a-b*r, a + b*r]   range\
y-axis: real interest rate(Rt), x-axis: Y_st

__(a)__ The Federal Reserve undertakes policy actions that have the effect of lowering the real interest rate below the marginal product of captial. _Hint: there might not be two IS for this question_

```python
a = 0
b = 3
r = 2
i = 0

Y_st = np.linspace(-a - b*r, a + b*r, 101)
R_t = (((a - Y_st[i]) / b) + r)

for i in range(0, 100):
    R_t = np.append(R_t, ((a - Y_st[i]) / b) + r) 
    
plt.plot(Y_st, R_t, "orange", label = "IS")
plt.title("When the real interest rate lower than MPK")
plt.xlabel("Short-run Outpt %(Y~t)")
plt.ylabel("Real interest rate %(Rt)")
plt.scatter(Y_st[50], R_t[50], s=50, c = "red", label = "initial")
plt.scatter(Y_st[60], R_t[60] ,s=50, c = "green", label = "below r")
plt.legend()
plt.show()
```

Becuase the real interest is endogenous variable, a point moves long the curve if the real interest rate changes. Because the real interest rate is lower than MPK = 2, short-run output is over 0 percent becuase people invest more.

## Q3) Imports and the Multiplier

The amount of goods that the US economy imports might depend on the current state of the economy as well as on potential GDP. For example, when the economy is booming, imports usually rise. To incorporate this channel into the model, suppose the import equation is given by 

\begin{align*}
\frac{IM_t}{\bar{Y_t}} = \bar{a_{im}} + \bar{n}\tilde{Y_{t}}.
\end{align*}

Assume the remainder of the model is unchanged from the original setup, as in Table 11.1 (pg. 283)

__(a)__ Derive and graph the IS curve for this new specification. Write your answers below

Y_st = (1 / (1 - n)) * (a - b*(Rt - r))\
R_t = ((Y_st * (1 - n) - a) / -b) + r

__(b)__ Assuming that the parameters for the IS curve are: $\bar{a}=0$, $\bar{b}=3$, $\bar{r}=2$, and $\bar{n}=2$. Plot both the original IS curve next to the new version of the IS curve from this question.

```python
a = 0
b = 3
r = 2
n = 0.2
i = 0

Y_st1 = np.linspace(-(a + b*r), a + b*r , 100)
Rt_8 = (a - Y_st1[i] / b) + r

for i in range(0, 99):
    Rt_8 = np.append(Rt_8, (a - Y_st1[i] / b) + r)

a = 0
b = 3
r = 2
n = 0.2
i = 0

Rt_im = ((a - (Y_st1[i] * (1 - n))) / b) + r

for i in range(0, 99):
    Rt_im = np.append(Rt_im, ((a - (Y_st1[i] * (1 - n))) / b) + r)
    
plt.plot(Y_st1, Rt_im, "orange", label = "new IS")
plt.plot(Y_st1, Rt_8, "green", label = "OG IS")
plt.title("With import Multiplier")
plt.xlabel("Short-run Output %(Y_st)")
plt.ylabel("real interest rate %(R_t)")
plt.legend()
plt.show()

print("(1 - n) / b is smaller than 1/ b so the slope is flatter")
```

## Q4) Consumption and the Real Interest Rate

According to the life-cycle/permanent-income hypothesis, consumption depends on the present dicounted value of income. An increase in the real interest rate will make future income worth less, thereby reducing the present discounted value and reducing consumption. To incorporate this channel into the model, suppose the consumption equation is given by 

\begin{align*}
C_t = \bar{a_c}\bar{Y_t}-\bar{b_c}(R_t-\bar{r})\bar{Y_t}
\end{align*}

Assume the remainder of the model is unchanged from the original setup, as in Table 11.1 (pg. 283)

__(a)__ Derive and graph the IS curve for this new specification. Write your answers below

Y_st = a -(b + b_c) * (Rt - r)\
Rt = ((a - Y_st) / (b + b_c)) + r\
becoming flatter slope because b became bigger

__(b)__ Assuming that the parameters for the IS curve are: $\bar{a}=0$, $\bar{b}_i=3$, $\bar{b}_c=1$, and $\bar{r}=2$. Plot both the original IS curve next to the new version of the IS curve from this question.

```python
a = 0
b = 3
b_c = 1
r = 2
i = 0

Y_st = np.linspace(-(a + (b+b_c)*r), a + (b+b_c)*r, 100)
R_to = ((a - Y_st[i]) / b) + r

for i in range(0, 99):
    R_to = np.append(R_to, ((a - Y_st[i]) / b) + r)
    
a = 0
b = 3
b_c = 1
r = 2
i = 0

R_tc = ((a - Y_st[i]) / (b + b_c)) + r
for i in range(0, 99):
    R_tc = np.append(R_tc, ((a - Y_st[i]) / (b + b_c)) + r)

plt.plot(Y_st, R_to, "orange", label = "OG IS")
plt.plot(Y_st, R_tc, "yellow", label = "C IS")
plt.title("Consumption and real interest rate")
plt.xlabel("Short-run Output")
plt.ylabel("real interest rate")
plt.legend()
plt.show()
```

__(c)__ How and why does it differ from the original IS curve? (_Hint:_ Think about the slope of the IS curve)

The slope is becoming flatter becuase (b + b_c) as a total b is becoming bigger.\
The slope 1 / (b + b_c) is smaller than 1 / b
