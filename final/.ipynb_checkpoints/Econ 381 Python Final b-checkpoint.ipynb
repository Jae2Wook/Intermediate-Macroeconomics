## Part 1: Solow-Romer

```python
## Parameter Values
l = 0.01
z = 0.02/l
n = 0.02  # why popualtion growth grew?
s = 0.1
delta = 0.06
alpha = 0.33



## Define functions
# 1) Steady state of capital in intensive units

kstar = lambda s, delta, z, l, n, alpha: ((s/(delta + z*l + n)) ** (1/ (1-alpha))) * (1-l)



# 2) Output function

output=lambda K,L,A,alpha,l: K ** alpha * (A* L* (1-l)) ** (1-alpha) 

## Definiting intial values for the path of Capital, Labor, and TFP.

K0 = 900
L0 = 500 
A0 = 1.7 

##Preallocating arrays

L = np.array([L0])
A = np.array([A0])
K = np.array([K0])
Y = np.array([output(K[0],L[0],A[0],alpha,l)])
C = np.array([(1-s)*Y])
t = np.array([0])
Delta_K= np.array([0])
k = np.array([K[0]/(A[0]*L[0])]) #Capital in intensive units k=K/AL
k_ss = kstar(s, delta, z, l, n, alpha)


## Loop parameters

i = 0 #Start counter
epsilon=0.01 # while loop tolerance

## The first loop simulates an economy reaching the steady state. The second loop extends the simulation for 10 periods
## after the steady state has been reached.


while (abs(k[i] - k_ss)) >epsilon:
    i += 1    
    t = np.append(t, i) #Time vector
    L = np.append(L, L[i-1] * (1 + n)) #Labor Supply Path
    A = np.append(A, A[i-1] * (1 + z*l)) #TFP Path
    Delta_K = np.append(Delta_K,s * output(K[i-1],L[i-1],A[i-1],alpha,l) - delta * K[i-1]) #Change in capital stock
    K= np.append(K,K[i-1] + Delta_K[i]) #Capital Stock Path
    k = np.append(k,K[i] / (A[i] * L[i])) #Path of Capital in intensive units
    Y = np.append(Y, output(K[i],L[i],A[i],alpha,l))
    C = np.append(C, (1-s) * output(K[i],L[i],A[i],alpha,l))

for i in range (i, i+10):
    i += 1    
    t = np.append(t, i)
    L = np.append(L, L[i-1] * (1 + n))
    A = np.append(A, A[i-1] * (1 + z*l))
    Delta_K = np.append(Delta_K,s * output(K[i-1],L[i-1],A[i-1],alpha,l) - delta * K[i-1])
    K = np.append(K,K[i-1] + Delta_K[i])
    k = np.append(k,K[i-1] / (A[i-1] * L[i-1]))
    Y = np.append(Y, output(K[i],L[i],A[i],alpha,l))
    C = np.append(C, (1-s) * output(K[i],L[i],A[i],alpha,l))
    

# Variable Steady State values at the end of the simulation. 
t[-1]
Ksteady = K[-1]
Asteady = A[-1]
Ysteady = Y[-1]
Lsteady = L[-1]
Csteady = C[-1]
Delta_Ksteady = Delta_K[-1]

# The economy will begin at the steady state for 10 periods and then face shocks to capital, output, and depreciation.

# Preallocating the variable (arrays) and setting initial values at the steady state from the previous exercise.
L = np.array([Lsteady])
A = np.array([Asteady])
K = np.array([Ksteady])
Y = np.array([Ysteady])
C = np.array([Csteady])
t = np.array([0])
Delta_K= np.array([Delta_Ksteady])
k = np.array([K[0]/(A[0]*L[0])]) 
k_ss = kstar(s, delta, z, l, n, alpha)
i = 0

# Letting the model follow the balanced growth path over 10 periods.
for i in range (0, 10):
    i += 1    
    t = np.append(t, i)
    L = np.append(L, L[i-1] * (1 + n))
    A = np.append(A, A[i-1] * (1 + z*l))
    Delta_K = np.append(Delta_K,s * output(K[i-1],L[i-1],A[i-1],alpha,l) - delta * K[i-1])
    K= np.append(K,K[i-1] + Delta_K[i])
    k = np.append(k,K[i-1] / (A[i-1] * L[i-1]))
    Y = np.append(Y, output(K[i],L[i],A[i],alpha,l))
    C = np.append(C, (1-s) * output(K[i],L[i],A[i],alpha,l))

# The path of variables in Venezuela for the past 5 years

# Period 11

n=0.013 #decrease
z=1.8   #decrease
delta=0.12 #increase

i = 11    
t = np.append(t, i)
L = np.append(L, L[i-1] * (1 + n))
A = np.append(A, A[i-1] * (1 + z*l))
Delta_K = np.append(Delta_K,s * output(K[i-1],L[i-1],A[i-1],alpha,l) - delta * K[i-1])
K= np.append(K,K[i-1] + Delta_K[i])
k = np.append(k,K[i-1] / (A[i-1] * L[i-1]))
Y = np.append(Y, output(K[i],L[i],A[i],alpha,l))
C = np.append(C, (1-s) * output(K[i],L[i],A[i],alpha,l))

# Period 12
n=0.008
z=1.5
delta=0.15

i = 12    
t = np.append(t, i)
L = np.append(L, L[i-1] * (1 + n))
A = np.append(A, A[i-1] * (1 + z*l))
Delta_K = np.append(Delta_K,s * output(K[i-1],L[i-1],A[i-1],alpha,l) - delta * K[i-1])
K= np.append(K,K[i-1] + Delta_K[i])
k = np.append(k,K[i-1] / (A[i-1] * L[i-1]))
Y = np.append(Y, output(K[i],L[i],A[i],alpha,l))
C = np.append(C, (1-s) * output(K[i],L[i],A[i],alpha,l))
    
    
# Period 13
n=0.00012
z=1.3
delta=0.18

i = 13    
t = np.append(t, i)
L = np.append(L, L[i-1] * (1 + n))
A = np.append(A, A[i-1] * (1 + z*l))
Delta_K = np.append(Delta_K,s * output(K[i-1],L[i-1],A[i-1],alpha,l) - delta * K[i-1])
K= np.append(K,K[i-1] + Delta_K[i])
k = np.append(k,K[i-1] / (A[i-1] * L[i-1]))
Y = np.append(Y, output(K[i],L[i],A[i],alpha,l))
C = np.append(C, (1-s) * output(K[i],L[i],A[i],alpha,l))
    

# Period 14

n=-0.0078
z=1.1
delta=0.2

i = 14    
t = np.append(t, i)
L = np.append(L, L[i-1] * (1 + n))
A = np.append(A, A[i-1] * (1 + z*l))
Delta_K = np.append(Delta_K,s * output(K[i-1],L[i-1],A[i-1],alpha,l) - delta * K[i-1])
K= np.append(K,K[i-1] + Delta_K[i])
k = np.append(k,K[i-1] / (A[i-1] * L[i-1]))
Y = np.append(Y, output(K[i],L[i],A[i],alpha,l))
C = np.append(C, (1-s) * output(K[i],L[i],A[i],alpha,l))

# Period 15

n=-0.015
z=1
delta=0.25

i = 15    
t = np.append(t, i)
L = np.append(L, L[i-1] * (1 + n))
A = np.append(A, A[i-1] * (1 + z*l))
Delta_K = np.append(Delta_K,s * output(K[i-1],L[i-1],A[i-1],alpha,l) - delta * K[i-1])
K= np.append(K,K[i-1] + Delta_K[i])
k = np.append(k,K[i-1] / (A[i-1] * L[i-1]))
Y = np.append(Y, output(K[i],L[i],A[i],alpha,l))
C = np.append(C, (1-s) * output(K[i],L[i],A[i],alpha,l))




# Simulating the path of variables towards the steady state as if things magically went back to normal in year 16.

epsilon=0.03 # Increasing the value of epsilon to reduce time toward the steady state.
z = 0.02/l
n = 0.02
delta = 0.06

while (abs(k[i] - k_ss)) >epsilon:
    i += 1    
    t = np.append(t, i)
    L = np.append(L, L[i-1] * (1 + n))
    A = np.append(A, A[i-1] * (1 + z*l))
    Delta_K = np.append(Delta_K,s * output(K[i-1],L[i-1],A[i-1],alpha,l) - delta * K[i-1])
    K= np.append(K,K[i-1] + Delta_K[i])
    k = np.append(k,K[i-1] / (A[i-1] * L[i-1]))
    Y = np.append(Y, output(K[i],L[i],A[i],alpha,l))
    C = np.append(C, (1-s) * output(K[i],L[i],A[i],alpha,l))
    
    
# How long did it take to get back to the steady state

print("Periods until Venezuela got back to the steady state:", t[i]-10)
    
# Simulating the economy for 10 periods after reaching the steady state.

for i in range (i, i+10):
    i += 1    
    t = np.append(t, i)
    L = np.append(L, L[i-1] * (1 + n))
    A = np.append(A, A[i-1] * (1 + z*l))
    Delta_K = np.append(Delta_K,s * output(K[i-1],L[i-1],A[i-1],alpha,l) - delta * K[i-1])
    K = np.append(K,K[i-1] + Delta_K[i])
    k = np.append(k,K[i-1] / (A[i-1] * L[i-1]))
    Y = np.append(Y, output(K[i],L[i],A[i],alpha,l))
    C = np.append(C, (1-s) * output(K[i],L[i],A[i],alpha,l))
    

# Paths of Capital in intensive units, Capital, Output, and Consumption
plt.plot(t,k)
plt.xlabel('Year')
plt.ylabel('k: Capital in Intensive Units')
plt.title('Path of k')
plt.show()

# Paths of Capital in Levels
plt.plot(t,K, label='Capital')
plt.xlabel('Year')
plt.ylabel('Output Units')
plt.title('Capital in Levels')
plt.legend()
plt.show()

# Paths of Labor in Levels
plt.plot(t,L, label='Labor')
plt.xlabel('Year')
plt.ylabel('Output Units')
plt.title('Labor in Levels')
plt.legend()
plt.show()


# Paths of Output and Consumption in Levels
plt.plot(t,Y, label='Output')
plt.plot(t,C, label='Consumption')
plt.xlabel('Year')
plt.ylabel('Output Units')
plt.title('Output and Consumption')
plt.legend()
plt.show()


# Paths of Capital in logs
plt.plot(t,K, label='Capital')
plt.xlabel('Year')
plt.ylabel('Log Output Units')
plt.title('Capital in Levels')
plt.legend()
plt.show()

# Paths of Labor in logs
plt.plot(t,L, label='Labor')
plt.xlabel('Year')
plt.ylabel('Log Output Units')
plt.title('Labor in Levels')
plt.legend()
plt.show()
```

## Part 2: AD-AS

```python
## Parameter Values paths

t = 20 # Each period is a quarter.
a = np.zeros(t) 
b = np.ones(t)/2
m = np.ones(t)
v = np.ones(t)/2
pi_bar = np.ones(t)*0.74 # Quarterly Inflation rate target - equivalent to 3% per year.
o = np.zeros(t)


# Preallocate loop variables

pi = np.zeros(t)
sr_y = np.zeros(t)

# Initial, steady state values for inflation and short run output

pi[0] = (pi_bar[0] + o[0] + (v[0]*b[0]*m[0])*pi_bar[0] + a[0]*v[0])/(1 + v[0]*b[0]*m[0])
sr_y[0] = a[0] -(b[0]*m[0])/(1 + v[0]*b[0]*m[0])*(pi[0] - pi_bar[0] + o[0] + a[0]*v[0])



# Shocks

a_i=np.array([-10, -5, -2, -1,  0]) # Sock to Investment
a_c=np.array([-10, -5, -2, -2, -2]) # Shock to Consumption
a_g=np.array( [15,  7,  0,  0,  0]) # Government stimulus
a[1:6] = a_i+a_c+a_g

o[1:2] = - 1 # One quarter oil shock
m[1:10]= 2 # Aggressive monetary policy to control inflation

# Inflation and short run output paths.

for i in range(1,t):
    
    pi[i] = (pi[i-1] + o[i] + (v[i]*b[i]*m[i])*pi_bar[i] + a[i]*v[i])/(1 + v[i]*b[i]*m[i])
    sr_y[i] = a[i] -(b[i]*m[i])/(1 + v[i]*b[i]*m[i])*(pi[i] - pi_bar[i] + o[i] + a[i]*v[i])
    
    
## Impulse response function for inflation and short run output.
print(a)

time = np.linspace(0,t-1,t)

plt.plot(time, pi , label = 'Inflation')
plt.plot(time, sr_y, label = 'Short-run Output')
plt.xlabel('Quarter')
plt.title('AD-AS Model - Inpulse response functions: Inflation and Short-run Output')
plt.xlim(0, time.size) 
plt.legend()
plt.show()
plt.show()
```

```python
## Simulate the path of an economy starting in a balanced growth path, P[0]=100, and Y[0]=100

# Setting up the variable paths

n = 5   # number of periods preceeding the shock
y_growth_rate = 0.004 #Output growth rate, quarterly 
p_bar_before = 0.0074 #Economy inflation rate, quarterly
p = np.zeros(t+n)
y = np.zeros(t+n)

# Initial values for the series

y[0] = 100
p[0] = 100

# The economy before the shock

for i in range(1,n):
    y[i]=y[i-1]*(1+y_growth_rate)
    p[i] = p[i-1]*(1 + p_bar_before)

    
# The economy after the shock.

for i in range(n-1,t+n):
    y[i] = y[i-1]*(1 + y_growth_rate + sr_y[i-n]/100)
    p[i] = p[i-1]*(1 + pi[i-n]/100)
    

    
## The path of output and prices over time    
    
    
time2 = np.linspace(0,t+n-1,t+n)

plt.plot(time2, p , label = 'Price Level')
plt.plot(time2, y, label = 'Output')
plt.xlabel('Quarter')
plt.title('AD-AS Model - Time Paths of the Price Level and Output')
plt.xlim(0, time2.size-1) 
plt.legend()
plt.show()


```

## Part 3: DGE

```python
#Parameter values

t = 15
alpha=1/3
beta= 0.98
delta=0.04

# Steady State Values

w_bar = lambda alpha, beta: (1-alpha) ** (1/(1-alpha)) * (beta/(1+beta)) ** (alpha/(1-alpha))
r_bar = lambda alpha, beta: (1+beta)/beta * alpha/(1-alpha)

# Initial Values for interest rate and wage vectors

r_0=r_bar(alpha,beta)
w_0=w_bar(alpha,beta)

# Functions

output = lambda K,L,alpha: K ** alpha * (L) ** (1-alpha)
wage_pres = lambda wage_past, alpha, beta: (1-alpha) * (beta/(1+beta) * wage_past) ** alpha 
int_pres = lambda wage_past, alpha, beta: alpha / ((beta/(1+beta) * wage_past) ** (1-alpha))


#Preallocating arrays and initial values.

w = np.array([w_0])
r = np.array([r_0])

c_1 = np.array ([w_0/(1+beta)])
c_2 = np.array([(1+r_0-delta)*beta/(1+beta)*w_0])
k_2 = np.array([beta/(1+beta)*w_0])
K = k_2
Y = output(K,1,alpha)

alpha= np.ones(t) * alpha 
beta = np.ones(t) * beta 
delta = np.ones(t) * delta 

# Sample shock, permanent decrease in the discount rate at period 8

beta[5:6] = 0.99
alpha[5:6]= .25


i = 0 #Start counter

for i in range(0,t-1):
    i += 1
    w = np.append(w, wage_pres(w[i-1],alpha[i],beta[i])) #Path of wages
    r = np.append(r, int_pres(w[i-1],alpha[i],beta[i])) #Path of interest rate
    k_2 = np.append(k_2,[beta[i]/(1+beta[i])*w[i-1]])
    c_1 = np.append(c_1,[w[i]/(1+beta[i])])
    c_2 = np.append(c_2,[(1+r[i]-delta[i])*beta[i]/(1+beta[i])*w[i]])
    Y = np.append(Y,output(k_2[i],1,alpha[i]))    
    

time=np.linspace(0,t-1,t)

plt.plot(time,w, label='Wages')
plt.plot(time,r, label='Interest Rate')
plt.xlabel('Year')
plt.ylabel('Output Units')
plt.title('The path of prices across time')
plt.legend()
plt.show()

plt.plot(time,k_2, label='Capital')
plt.plot(time,np.ones(t), label='Labor')
plt.plot(time,Y, label='Output')
plt.xlabel('Year')
plt.ylabel('Output Units')
plt.title('The path of the economy across time')
plt.legend()
plt.show()

plt.plot(time,c_1, label='Consumption when Young')
plt.plot(time,c_2, label='Consumption when Old')
plt.xlabel('Year')
plt.ylabel('Output Units')
plt.title('The path of household consumption')
plt.legend()
plt.show()
```

### Fundamental Changes to the Economy

These two pressures will likely lead to an increase in savings and the share of income that is accrued by households.

#### Model the fundamental changes in the economy using the DGE model.

a) How do the events previously describe translate to parameter values in the DGE model? Justify your choices.

Increasing the household discount rate will lead to higher level of savings. I increased the discount rate from 0.98 to 0.99 for one generations. At the same time the increase in worker's barganing power means a decrease in alpha for the same lenght of time. 

b) Analyze the path of the economy over time as it responds to the changes you made in part a). 

As expected you see an increase in savings by households. That accompanies an increase in capital. Because of the increase of capital we observe a decrease in the marginal product of capital. In equilibrium, the marginal product of capital is equal to the interest rate. The share of income going to households goes from 66% to 75%. This leads to higher levels of consumption and savings despite a $\underline{decrease in the interest rate.}$(Is it because higher wage makes the real interest rate low?) Because households provide the same level of labor each period, the increase in capital raises output. More output can be split into more consumption in both periods.

c) Assume the changes in the parameters of the model only affect one generation. How many generations will live with the decisions of their predecesors?

The consequence of a one-period change in parameters affects the consumption of the young for 4 generations. That is due to the increase level of savings. The only variable that has persistence in this model is capital. Changes in the capital stock affect output for extended periods of time. d) The DGE model we've worked with is only the most basic model of this type. Suggest one possible extension to the model and explain how that extension might change your results from part a) and b)

There are lots of possibilities. Here is a short list:
    
-Infinite horizon agents (or at least agents with longer lives)

-Decisions about labor supply (do I work? if so, how much?)

-More dimensions of heterogeneity (skills, occupations, different kinds of capital) 

-Different sources of uncertainty (can't see the future perfectly)

-Price frictions, wage frictions, information frictions

-Game theoretic/strategic interactions within and between households and firms 

-Borrowing constraints 

-Monetary policy 

-Fiscal policy 

-Political economy 

-Financial markets
