__(a)__: Using Python, create two time series plots showing the inflation rate for China and India for the years given in the data. 

```python
# Import the data sets
china_url = "https://raw.githubusercontent.com/dylanjm/econ_381_development/master/data/ch08_china_inflation.csv"
china = pd.read_csv(china_url,header=0, names=['Date', 'China'])

india_url = "https://raw.githubusercontent.com/dylanjm/econ_381_development/master/data/ch08_india_inflation.csv"
india = pd.read_csv(india_url, header = 0, names = ['Date', 'India'])

#Convert the date variable to a datetime object
ChinaDate = np.array(china['Date'], dtype = np.datetime64)
ChinaInflation = np.array(china['China'])

IndiaDate = np.array(india['Date'], dtype = np.datetime64)
ChinaInflation = np.array(india['India'])

#plot
plt.plot(ChinaDate, china['China'], label ='China')
plt.xlabel('Years')
plt.ylabel('Inflation Rate')
plt.title('Inflation of China')
plt.legend()
plt.show()

plt.plot(IndiaDate, india["India"], label = 'India')
plt.xlabel('Years')
plt.ylabel('Inflation Rate')
plt.title('Inflation of India')
plt.legend()
plt.show()
```

## Q3) Inflation and interest rates in Italy (FRED Question)

Using the data found [Here](https://github.com/dylanjm/econ_381_development/tree/master/data) import the data for `ch08_italy_inflation.csv` and `ch08_italy_bond.csv`. Ensure to use the __raw url__ for your Python code. 

__(a)__: Using Python, create a __single__ plot showing both the 10 year bond yields and inflation rate for Italy. Use the dates provided by the data. 

```python
# Python Code Here:

italy_url = "https://raw.githubusercontent.com/dylanjm/econ_381_development/master/data/ch08_italy_bond.csv"
italy_url2 = pd.read_csv("https://raw.githubusercontent.com/dylanjm/econ_381_development/master/data/ch08_italy_inflation.csv", header =0, names = ['Date','Inflation'])
italy = pd.read_csv(italy_url, header = 0, names=['Date', 'Bond'])
italy['Date'] = pd.to_datetime(italy['Date'])
italy['Year'] = italy['Date'].dt.year
ItalyBond = np.array(italy['Bond'])
plt.plot(italy['Year'], italy['Bond'], label = 'Bond')
plt.plot(italy['Year'], italy_url2['Inflation'], label = 'Inflation')
plt.xlabel('Year')
plt.ylabel('Interest Rate')
plt.title('Inflation rates & Bonds')
plt.legend(loc = 'upper center')
plt.show()
```

__(b)__: Why do these two series tend to move together over time?

If the inflation rises than the and bonds' interest which is fixed, who holds bonds losses money by inflation.
If the inflation falls than the bonds' interest which is fixed, who holds bonds is better-off.
To manage these issue, bonds interest rate should change as much as inflation changes.
The bonds' interest rates are moving by the expectation of the economy of Italy.
Bonds' interest rate is higher than the inflation rate because the country wants to attract the people from making money.

__(c)__: What economic concept corresponds to the difference between these two series?

Fisher equation. Same as Fisher equaiton (nominal - real = i), difference interest rate between the Bonds and inflation is the gain from the Bonds.
